# Task 2 :
## Real Estate Price Distribution

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
file_path = 'C:\\Users\\HP\\Downloads\\realtor-data.zip.csv'
dataf = pd.read_csv(file_path)
```

```python
dataf.columns = map(str.lower,dataf.columns)
```

```python
X = dataf.head(2000)
```

```python
X.head(10)
```

```python
plt.figure(figsize=(8,10))
sns.scatterplot(x="price",y="house_size",data=X,hue="bed",alpha = 0.5 ,palette = "inferno")
plt.xlabel("PRICE")
plt.ylabel("HOUSE SIZE")
plt.title("Real Estate Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(20,20))
sns.boxplot(x="house_size", y="city", data=X , orient = 'v')
plt.xlabel("PRICE")
plt.ylabel("CITY")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.histplot(X['price'],bins=30, kde=True)
plt.xlabel("PRICE")
plt.ylabel("FREQUENCY")
plt.title("HISTOGRAM")
plt.show()
```
