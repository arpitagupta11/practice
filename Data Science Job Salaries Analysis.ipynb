# Data Science job Salary Analysis

## Overview

This notebook provides an in-depth analysis of data science salaries using a comprehensive dataset. The goal is to uncover trends, compare salary distributions, and gain actionable insights into the factors influencing data science job salaries.

### Dataset

The dataset used for this analysis can be found on [Kaggle](https://www.kaggle.com/datasets). It includes various features related to data science job salaries, such as:

- `work_year`: The year of the salary data.
- `job_title`: The job title of the employee.
- `experience_level`: The level of experience of the employee (e.g., Junior, Mid-level, Senior).
- `employee_residence`: The location where the employee resides.
- `remote_ratio`: The ratio of remote work.
- `company_size`: The size of the company.
- `salary_in_usd`: The salary of the employee in USD.

## Setup

Before starting the analysis, we need to import the necessary libraries.

```python
import pandas as pd

````python
from matplotlib import pyplot as plt
import pandas as pd 
import seaborn as sns
````

````python
#load the dataset 

df = pd.read_csv("/kaggle/input/data-science-jobs-salaries-csv/Data Science Jobs Salaries.csv")
````

````python
df.head(2)
````

````python
#check if there are any missing values
df.isnull().sum()
````

````python
# Identify unique values in the 'work_year' column to check for invalid entries
print(df['work_year'].unique())
````

````python
# Replace invalid entries with a valid year (e.g., '2021')
# You can choose to replace or drop these rows based on your requirement
df['work_year'] = df['work_year'].replace('2021e', '2021')

# Convert the 'work_year' column to integer type
df['work_year'] = df['work_year'].astype(int)
````

````python
# Display summary statistics
print(df.describe())
````

````python
df.head(5)
````

````python
# Analyze salary trends over the years
salary_trends = df.groupby('work_year')['salary_in_usd'].mean().reset_index()

# Plot salary trends over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=salary_trends, x='work_year', y='salary_in_usd')
plt.title('Trends in Data Science Salaries Over Years')
plt.xlabel('Year')
plt.ylabel('Average Salary in USD')
plt.show()
````

````python
salary_trends
````

````python
# Compare salary distributions across job titles
plt.figure(figsize=(14, 7))
sns.boxplot(data=df, x='job_title', y='salary_in_usd')
plt.title('Salary Distributions Across Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Salary in USD')
plt.xticks(rotation=90)
plt.show()
````

````python
# Impact of experience on salaries
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='experience_level', y='salary_in_usd')
plt.title('Impact of Experience on Salaries')
plt.xlabel('Experience Level')
plt.ylabel('Salary in USD')
plt.show()
````

````python
# Impact of employee location on salaries
plt.figure(figsize=(14, 7))
sns.barplot(data=df, x='employee_residence', y='salary_in_usd')
plt.title('Impact of Employee Location on Salaries')
plt.xlabel('Employee Residence')
plt.ylabel('Salary in USD')
plt.xticks(rotation=90)
plt.show()
````

````python
# Impact of remote work on salaries
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='remote_ratio', y='salary_in_usd')
plt.title('Impact of Remote Work on Salaries')
plt.xlabel('Remote Ratio')
plt.ylabel('Salary in USD')
plt.show()
````

````python
# Impact of company size on salaries
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='company_size', y='salary_in_usd')
plt.title('Impact of Company Size on Salaries')
plt.xlabel('Company Size')
plt.ylabel('Salary in USD')
plt.show()
````

````python
# Plot the impact of company size on salaries using a boxplot
plt.figure(figsize=(14, 8))
sns.set(style="whitegrid")
boxplot = sns.boxplot(data=df, x='company_size', y='salary_in_usd', palette="viridis")
plt.title('Impact of Company Size on Data Science Salaries', fontsize=16)
plt.xlabel('Company Size', fontsize=14)
plt.ylabel('Salary in USD', fontsize=14)

# Enhance the x-axis labels for readability
plt.xticks(ticks=[0, 1, 2], labels=['Small', 'Medium', 'Large'], fontsize=12)


plt.show()
````

````python
g = sns.FacetGrid(df, col="experience_level", hue="job_title", height=4, aspect=1)
g.map(sns.lineplot, "work_year", "salary")
g.add_legend()
plt.title('Salary Trends by Experience Level')
plt.xlabel('Year')
plt.ylabel('Salary (USD)')
plt.show()
````
