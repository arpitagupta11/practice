# Customer Purchase Behavior Analysis

## Introduction

In this notebook, we analyze a dataset containing customer purchase behavior to gain insights into purchasing patterns and segment customers based on their behavior. This analysis involves data cleaning, exploratory data analysis (EDA), and visualization to understand relationships and identify trends.

## Dataset Description

The dataset `shopping_behavior_updated.csv` contains information about customer purchases. Key features include:

- **Purchase Amount (USD)**: The total amount spent in a transaction.
- **Frequency of Purchases**: How often the customer makes a purchase (e.g., Annually, Quarterly, Fortnightly, Weekly).
- **Gender**: The gender of the customer.
- **Item Purchased**: The specific item purchased.
- **Category**: The category to which the item belongs.
- **Location**: The location where the purchase was made.
- **Size**: The size of the purchased item.
- **Color**: The color of the purchased item.
- **Season**: The season during which the purchase was made.
- **Subscription Status**: Whether the customer has a subscription.
- **Shipping Type**: The type of shipping selected.
- **Discount Applied**: Whether a discount was applied to the purchase.
- **Promo Code Used**: Whether a promo code was used.
- **Payment Method**: The payment method used for the purchase.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Load the dataset
file_path = "C:\\Users\\hp\\Downloads\\shopping_behavior_updated.csv"
data = pd.read_csv(file_path)

# Display the first 8 rows of the dataset
print(data.head(2))
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(data['Purchase Amount (USD)'], data['Frequency of Purchases'], alpha=0.7)
plt.xlabel('Purchase Amount (USD)')
plt.ylabel('Frequency of Purchases')
plt.title('Scatter Plot of Purchase Amount vs. Frequency of Purchases')
plt.grid(True)
plt.show()
````

````python
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.countplot(x='Frequency of Purchases', data=data)
plt.xlabel('Frequency of Purchases')
plt.ylabel('Count')
plt.title('Bar Plot of Frequency of Purchases')
plt.grid(True)
plt.show()
````

````python
plt.figure(figsize=(12, 8))
sns.countplot(y='Gender', data=data, order=data['Gender'].value_counts().index)
plt.xlabel('Count')
plt.ylabel('Gender')
plt.title('Count Plot of Gender')
plt.grid(True)
plt.show()
````

````python
plt.figure(figsize=(12, 6))
data.groupby('Frequency of Purchases')['Purchase Amount (USD)'].mean().plot(kind='line', marker='o')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Average Purchase Amount (USD)')
plt.title('Line Plot of Average Purchase Amount by Frequency of Purchases')
plt.grid(True)
plt.show()
````

````python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Frequency of Purchases', y='Purchase Amount (USD)', data=data, palette='Accent')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Purchase Amount (USD)')
plt.title('Box Plot of Purchase Amount by Frequency of Purchases')
plt.grid(True)
plt.show()
````

````python
sns.pairplot(data[['Purchase Amount (USD)', 'Frequency of Purchases']])
plt.suptitle('Pair Plot of Purchase Amount and Frequency of Purchases', y=1.02)
plt.show()
````

````python
plt.figure(figsize=(12, 8))
sns.violinplot(x='Frequency of Purchases', y='Purchase Amount (USD)', data=data, palette='inferno')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Purchase Amount (USD)')
plt.title('Violin Plot of Purchase Amount by Frequency of Purchases')
plt.grid(True)
plt.show()
````

````python
g = sns.FacetGrid(data, col='Gender', margin_titles=True)
g.map_dataframe(sns.scatterplot, x='Purchase Amount (USD)', y='Frequency of Purchases')
g.set_axis_labels('Purchase Amount (USD)', 'Frequency of Purchases')
g.set_titles(col_template="{col_name}")
plt.show()
````

````python
plt.figure(figsize=(10, 6))
sns.histplot(data['Purchase Amount (USD)'], bins=30, kde=True)
plt.xlabel('Purchase Amount (USD)')
plt.ylabel('Frequency')
plt.title('Histogram of Purchase Amount (USD)')
plt.grid(True)
plt.show()
````
