```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

# Line plot


```python
#matplotlib
var = [1,2,3,4,5,6,7]
var_1 = [8,9,10,11,1,2,3]
plt.plot(var,var_1)
plt.show()
```

```python
#seaborn
x_1 = pd.DataFrame({
    "var": var,
    "var_1" : var_1
})

sns.lineplot(x = "var" , y = "var_1", data = x_1)
plt.show()
```

```python
# seaborn if you dont want to give dictionary n all then 
var = [1,2,3,4,5,6,7]
var_1 = [8,9,10,11,1,2,3]

sns.lineplot(x=var,y=var_1)

plt.show()
```

# new dataset


```python
dataf = sns.load_dataset("penguins")
dataf
```

```python
sns.lineplot(x="island", y ="bill_depth_mm", data = dataf , hue = "sex", style = "sex" , palette="inferno", markers =["o" , ">"] , dashes = False , legend = "full")# , size =10)
plt.grid()
plt.title("PENGUIN INFO")
plt.show()
```

## seaborn barplot


```python
order_1 = ["Biscoe","Dream" , "Torgersen"]
sns.barplot(x = "island", y="bill_length_mm", data= dataf , hue = "sex" ,
            palette = "Accent" , order = order_1 ,hue_order=["Female","Male"], orient = "v", saturation=20, )
           #ci=80, n_boot = 5, ,capsize=1
plt.show()
```

```python
order_1 = ["Biscoe","Dream" , "Torgersen"]
sns.barplot(x = "bill_depth_mm", y="bill_length_mm", data= dataf , orient = "v", palette = "inferno")
           #ci=80, n_boot = 5, 
plt.show()
```

# scatter plot

```python
sns.scatterplot(x = "bill_depth_mm", y="bill_length_mm", data= dataf ,hue = "sex", palette = "Accent", style="sex",size="sex",sizes=(100,60),alpha = 1)
plt.show()
```

# heat map

```python

# Select only numerical columns
numerical_data = dataf[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]

# Create the heatmap
sns.heatmap(data=numerical_data.corr(), annot=True, cmap='coolwarm')
plt.show()
```

# count plot

```python
sns.countplot(x = "island" , data= dataf , palette = "inferno" , hue = "sex" )
plt.show()
```

# violin plot

```python
dataf = sns.load_dataset("penguins").head(50)

sns.violinplot(x = "island", y="bill_length_mm", data= dataf, hue = "sex")
plt.show()
```

# pair plot

```python
sns.pairplot(dataf,vars =["bill_length_mm","bill_depth_mm"] ,hue = "sex" , kind="reg")
plt.show()
```

```python
sns.pairplot(dataf,hue = "sex" , kind = "reg") # "scatter","reg","hist","kde"
plt.show()
```

# strip plot

```python
dataf = sns.load_dataset("penguins")
sns.stripplot(x="island",y="bill_length_mm",data=dataf, hue = "sex")
plt.show()
```

# Box plot

```python
sns.set(style = "whitegrid")
sns.boxplot(x="island",y="bill_length_mm",data=dataf, hue = "sex" , color = "b" , showmeans = True,linewidth=2)
plt.show()
```

# cat plot

```python
sns.catplot(x="island", y="bill_length_mm", data=dataf,hue="sex", kind="swarm") #'strip', 'swarm', 'box', 'boxen', 'violin', 'bar', 'count', and 'point'.
plt.show()
```
